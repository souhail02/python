```python
dataset= data1.values.tolist()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

# Convertir toutes les valeurs de la liste de transactions en chaînes de caractères
dataset = [[str(item) for item in transaction] for transaction in dataset]

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)  # Appliquer le codage one-hot sur notre ensemble de données
data = pd.DataFrame(te_ary, columns=te.columns_)  # Créer un nouveau DataFrame à partir de notre tableau Numpy

# Afficher le DataFrame résultant
print(data)
```

```python
from mlxtend.frequent_patterns import apriori
apriori(data, min_support=0.1)
```

```python
frequent_itemsets=apriori(data, min_support=0.1, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=0.7)
```
