```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering


features_for_clustering = ['PURCHASES', 'CREDIT_LIMIT']
X_clustering = data1[features_for_clustering]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clustering)

linkage_matrix = linkage(X_scaled, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()

num_clusters = 5
hierarchical_model = AgglomerativeClustering(n_clusters=num_clusters, affinity='euclidean', linkage='ward')
hierarchical_model.fit(X_scaled)

data1['Cluster'] = hierarchical_model.labels_

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PURCHASES', y='CREDIT_LIMIT', hue='Cluster', data=data1, palette='viridis')
plt.title('Clusters based on PURCHASES and CREDIT_LIMIT')
plt.xlabel('PURCHASES')
plt.ylabel('CREDIT_LIMIT')
plt.show()
```
